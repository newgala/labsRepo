```python
mean_lastclaim=home3['months since last claim'].mean()
mean_complaints=home3['number of open complaints'].mean()
```

```python
home3['months since last claim']=home3['months since last claim'].fillna(mean_lastclaim)
home3['number of open complaints']=home3['number of open complaints'].fillna(mean_complaints)
```

```python
home3['state']= home3['state'].fillna("Unknown")
home3['response']=home3['response'].fillna("Unknown")
home3['vehicle class']=home3['vehicle class'].fillna("Unknown")
home3['vehicle size']=home3['vehicle size'].fillna("Unknown")
home3['vehicle type']=home3['vehicle type'].fillna("Unknown")
```
